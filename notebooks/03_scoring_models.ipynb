# SignalFire Scoring Models

Inspect company intelligence profiles, opportunity scores, momentum, and readiness outputs.

```python
from pathlib import Path
import duckdb

DB_PATH = Path('../data/processed/signalfire.duckdb')
assert DB_PATH.exists(), 'Run the pipeline before opening this notebook.'
con = duckdb.connect(str(DB_PATH), read_only=True)
```

```python
con.execute('SELECT * FROM company_profiles ORDER BY "Consulting Opportunity Score" DESC LIMIT 20').df()
```

```python
con.execute('''
SELECT company, industry, theme, strategic_priority_score, strategic_momentum_score,
       priority_growth_rate, transformation_readiness_score, consulting_opportunity_score
FROM scorecard
ORDER BY consulting_opportunity_score DESC
LIMIT 25
''').df()
```

```python
con.execute('''
SELECT industry, theme, AVG(consulting_opportunity_score) AS avg_opportunity
FROM scorecard
GROUP BY industry, theme
ORDER BY avg_opportunity DESC
LIMIT 25
''').df()
```
