# SignalFire NLP Analysis

Explore theme classification, TF-IDF keywords, named entities, and discovered topics.

```python
from pathlib import Path
import duckdb

DB_PATH = Path('../data/processed/signalfire.duckdb')
assert DB_PATH.exists(), 'Run the pipeline before opening this notebook.'
con = duckdb.connect(str(DB_PATH), read_only=True)
```

```python
con.execute('''
SELECT company, industry, year, theme, theme_intensity
FROM company_theme_scores
ORDER BY theme_intensity DESC
LIMIT 25
''').df()
```

```python
con.execute('''
SELECT ticker, company, term, AVG(tfidf_weight) AS avg_weight
FROM keywords
GROUP BY ticker, company, term
ORDER BY avg_weight DESC
LIMIT 30
''').df()
```

```python
con.execute('''
SELECT topic_id, STRING_AGG(term, ', ' ORDER BY weight DESC) AS terms
FROM topics
GROUP BY topic_id
ORDER BY topic_id
''').df()
```
