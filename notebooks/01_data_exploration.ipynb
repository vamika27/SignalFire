# SignalFire Data Exploration

Use this notebook to inspect real SEC filings, earnings-call transcripts, and Layoffs.fyi inputs after running the pipeline.

```python
from pathlib import Path
import duckdb
import pandas as pd

DB_PATH = Path('../data/processed/signalfire.duckdb')
assert DB_PATH.exists(), 'Run: python -m signalfire.src.pipeline --max-filings 2'
con = duckdb.connect(str(DB_PATH), read_only=True)
```

```python
con.execute('SHOW TABLES').df()
```

```python
documents = con.execute('SELECT * FROM documents LIMIT 20').df()
documents
```

```python
con.execute('''
SELECT source, industry, COUNT(*) AS documents
FROM documents
GROUP BY source, industry
ORDER BY documents DESC
''').df()
```
